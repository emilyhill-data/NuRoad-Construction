# 04 Clean the Data
## a. check for mixed variables & change if needed
## b. check for duplicates
## c. check for nulls/blanks
## d. check for unclear data/entry adjustments

```python
# a. check for mixed variables
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
        print (col)
```

```python
#change all mixed variables to object/string
df['City'] = df['City'].astype('str')
df['Timezone'] = df['Timezone'].astype('str')
df['Wind_Direction'] = df['Wind_Direction'].astype('str')
df['Sunrise_Sunset'] = df['Sunrise_Sunset'].astype('str')
df['Street'] = df['Street'].astype('str')
df['Civil_Twilight'] = df['Civil_Twilight'].astype('str')
df['Nautical_Twilight'] = df['Nautical_Twilight'].astype('str')
df['Astronomical_Twilight'] = df['Astronomical_Twilight'].astype('str')
df['Zipcode'] = df['Zipcode'].astype('str')
df['Airport_Code'] = df['Airport_Code'].astype('str')
df['Weather_Condition'] = df['Weather_Condition'].astype('str')
df['Weather_Timestamp'] = df['Weather_Timestamp'].astype('str')
```

```python
#check to make sure no mixed variable types
for col in df.columns.tolist():
    weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df[weird]) > 0:
        print (col)
```

```python
# b. check for duplicates by creating a duplicates sub-dataset
df_dups = df[df.duplicated()]
```
