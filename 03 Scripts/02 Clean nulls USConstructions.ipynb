## b. Create df3 without records where Distance(mi) is zero

```python
 #create a df with all values where the distance = 0
df_miss_dist = df2.loc[df2['Distance(mi)'] == 0]
```

```python
# some of the missing data is due to not having an "end" lat/long, others were just not measured.
# this data consists of 374,171 records and represents approximately 6.06% of the original dataset
df_miss_dist.describe()
```

```python
#create new dataframe with only valid distances
#cumulatively, will have removed 489,565 records which represents 7.93% of original data
df3 = df2.loc[df2['Distance(mi)'] != 0]
```

## d. create df4 with null Visibility(mi) records omitted

```python
df3['Visibility(mi)'].value_counts(dropna = False).sort_index()
#since visibility only has 40,038 null records, I will delete these for my analysis
```

```python
# create a new df with not-null values for visibility
# cumulatively, I have removed 529,603/6,170,627 records, which is 8.58% of total records
Vis_nomt = pd.notnull(df3['Visibility(mi)'])
df4 = df3[Vis_nomt]
```
